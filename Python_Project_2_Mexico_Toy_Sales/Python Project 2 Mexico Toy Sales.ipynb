##### Grouping Store City by Stock On Hand with Ranking

```python
def Ranking(x):
    if x < 500:
        return 'Low Stock'
    elif x < 1000:
        return 'Med Stock'
    elif x < 1500:
        return 'Med-High Stock'
    else:
        return 'High Stock'



StoreCityStockOnHandSorted["Ranking"] = StoreCityStockOnHandSorted["Stock On Hand"].apply(Ranking)
```
